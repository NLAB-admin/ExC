"This is a very simple way to interact with the API. The main thing to know is that the messages list has a dictionary with 2 keys: roles and content.
The content is simply the content of the message, while there are three main roles : “system”, “user”, or “assistant”. The “user” is the one who gives the instructions and is being used in the code above.
[One could] add the system role to set the behavior of the assistant and the assistant role to store prior responses."


```python
content = input("User: ")
messages = [
 {"role": "system", "content" : "You’re a kind helpful assistant"}
]
messages.append({"role": "user", "content": content})

completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=messages
)

chat_response = completion.choices[0].message.content
print(f'ChatGPT: {chat_response}')
```

Here are some extra parameters:

- max_token: The maximum number of tokens to generate in the completion (here you can see the tokenizer that OpenAI uses)
- temperature: The sampling temperature to use. Values close to 1 will give the model more risk/creativity, while values close to 0 will generate well-defined answers.
- n: The number of chat completion choices to generate for each input message.

To see all the parameters available, go to the official [documentation](https://platform.openai.com/docs/api-reference/chat/create).

```python
def create_panel_prompt(product, gender, age, location, indrole):
  prompt = f"""In your opinion how do you think the typical UK consumer would feel about using/consuming a {product}. 
  Please provide a rating for your feelings as:
  1) Don't enjoy at all
  2) Don't enjoy
  3) Neither enjoy or don't enjoy
  4) Enjoy
  5) Enjoy very much
  
  Please assume the consumer can afford the product or service.
  You are a {indrole} ({gender}, {age} years old) living in {location}. 
  What is your rating (1, 2, 3, 4, or 5)?
  Please only provide a number without explanation."""
  return prompt
```

```python
location = ["London", "Manchester", "Birmingham", "Leeds", "Glasgow", 
"Liverpool", "Newcastle", "Sheffield", "Bristol", "Belfast", 
"Edinburgh", "Southampton", "Brighton", "Cardiff", "Leicester", 
"Coventry", "Hull", "Bradford", "Stoke-on-Trent", "Wolverhampton"]
age = [35, 40, 45, 50]
gender = ["female","male"]
indrole = ["Chief Data Officer", "Professor"]
product_panel = ["holiday", "car", "savings account", "present for a close relative", "energy supplier", "takeout lunch in a location you are unfamiliar with", "bedroom furniture", "new smartphone", "clothes for everyday wear", "food and drink items for the main grocery shop", "wine", "music concert", "home security system", "toothpaste", "spa", "video game console", "clothes for a friend's wedding", "running shoes", "coffee machine", "jewellery"]
quest = ["using/consuming", "searching for information prior purchase regarding"]
```

```python
import itertools
#prompt_panel = list(itertools.product(gender, age, location, indrole))
prompt_panel = list(itertools.product(gender, age, indrole))
#prompt_panel[0][1]
len(prompt_panel)
```

```python
# create empty dataframe
import pandas as pd 
# data_prompt = pd.DataFrame(columns=['id','question','product','rating','gender','age','location','indrole'])
# NEVER grow a DataFrame row-wise!
# create product/service list + question
prompt_quest = list(itertools.product(quest, product_panel))
len(prompt_quest)
```

```python
def create_panel_prompt2(quest, product, gender, age, location, indrole):
  prompt = f"""In your opinion how do you think the typical UK consumer would feel about {quest} a {product}. 
  Please provide a rating for your feelings as:
  1) Don't enjoy at all
  2) Don't enjoy
  3) Neither enjoy or don't enjoy
  4) Enjoy
  5) Enjoy very much
  
  Please assume the consumer can afford the product or service.
  You are a {indrole} ({gender}, {age} years old) living in {location}. 
  What is your rating (1, 2, 3, 4, or 5)?
  Please only provide a number without explanation."""
  return prompt
```

```python
def create_panel_prompt3(quest, product, gender, age, indrole):
  prompt = f"""You are a {indrole} ({gender}, {age} years old) living in UK. 
  In your opinion how do you think the typical UK consumer would feel about {quest} a {product}?
  Please assume the consumer can afford the product or service.
  Please provide a rating on behalf of the typical UK consumer:
  Don't enjoy at all = 1
  Don't enjoy = 2
  Neither enjoy or don't enjoy = 3
  Enjoy = 4
  Enjoy very much = 5
  Please provide only a number without explanation.
  """
  return prompt
```

We need now to loop through scenarios to get a rating number for each 'fake' subject for each scenario

```python
lid = []
lqu = []
lpr = []
lrt = []
lge = []
lag = []
llo = []
lro = []

for i in [0, 1, 2, 3, 4, 5]:
  #range(len(prompt_panel))
  # these are fake subjects
  id = i+1
  for j in [0, 1, 2, 3, 4, 5]:
    #range(len(prompt_quest))
    lid.append(id)
    lge.append(prompt_panel[i][0])
    lag.append(prompt_panel[i][1])
    llo.append(prompt_panel[i][2])
    lro.append(prompt_panel[i][3])
    if prompt_quest[j][0] == "using/consuming":
      q = 1
    else:
      q = 2
    lqu.append(q)
    lpr.append(prompt_quest[j][1])
    prompt = create_panel_prompt2(prompt_quest[j][0], prompt_quest[j][1], prompt_panel[i][0], prompt_panel[i][1], prompt_panel[i][2], prompt_panel[i][3])
    panel_prompt2 = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "user", "content": prompt}
      ]
    )
    lrt.append(panel_prompt2.choices[0].message.content)
```
